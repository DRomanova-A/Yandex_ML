Импортируем модули

```python
from scipy.stats import t
import numpy as np
import matplotlib.pyplot as plt
```

Сгенерируем выборку объёма $n=1000$:

```python
fig, ax = plt.subplots(1, 1) # Создание объекта Figure и области рисования

n = 1000 # Задание выборки

x = np.linspace(t.ppf(0.01, n), t.ppf(0.99, n), 1000) # Задаем интервал с количеством элементов  
rv = t(n) # Задание распределения Стьюдента
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='t распределение') # Построение графика распределения

r = t.rvs(n, size=1000) # Генерируем выборку

ax.hist(r, density=True, histtype='bar', alpha=0.2, ec='black') # Построение гистограммы
ax.legend(loc='best', frameon=False) # Выводим название распределения
plt.ylabel('f(y)') # Присвоение названия вертикальной оси 
plt.xlabel('$y$') # Присвоение названия горизонтальной оси 
plt.show() # Отображение графика
```

```python
from scipy.stats import norm
fig, ax = plt.subplots(1, 1)
n = 5

x = np.linspace(norm.ppf(0.01),norm.ppf(0.99), 100)

rv = norm(0,np.sqrt(1/(n-2)))
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='распределение')

r = rv.rvs(size=1000)

ax.hist(r, density=True, histtype='bar', alpha=0.2, ec='black')
ax.legend(loc='best', frameon=False)
plt.ylabel('f(y)')
plt.xlabel('$y$')
plt.show()
```
